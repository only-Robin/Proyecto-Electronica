```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Para filtrar.csv')
data.drop('Unnamed: 0',axis = 1, inplace = True)
data_lluv = data[data.Soleado ==1]
data_lluv[data.Hora =='9:00'].describe()
```

```python
promedios = []
desv = []
horas = data.Hora.unique()

for h in horas:
    promedios.append(data[data.Hora == h]['Irradiacion [Wh/m2]'].mean())
    desv.append(data[data.Hora == h]['Irradiacion [Wh/m2]'].std())

estats = pd.DataFrame({'Hora':horas,'Promedio':promedios,'Desviacion':desv})
estats.set_index('Hora',inplace = True)
estats
```
