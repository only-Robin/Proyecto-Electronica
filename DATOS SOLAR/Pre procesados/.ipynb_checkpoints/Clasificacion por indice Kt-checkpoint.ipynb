### Indice Kt por dia

```python
indices = []
fecha = datos.Fecha.unique()
for f in fecha:
    real = datos[datos.Fecha == f]['Irradiacion [Wh/m2]'].values
    extraterrestre = datos[datos.Fecha == f]['Irradiacion extraterrestre [Wh/m2]'].values
    suma = 0
    j= 0
    for i in range(1,len(real)):
        if extraterrestre[i] == 0:
            pass
        else:
            suma += real[i]/extraterrestre[i]
            j+=1
    indice_actual = suma/j
    #print(indice_actual)
    indices.append(indice_actual)
```

```python
indices = pd.DataFrame({'Fecha':fecha, 'Indice':indices})
```

```python
indices_fecha = indices.set_index('Fecha')
indices_fecha.min()
```
