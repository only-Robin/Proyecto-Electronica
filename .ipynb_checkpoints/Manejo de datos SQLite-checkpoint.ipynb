```python
import sqlite3
import datetime as dt
import pandas as pd
import numpy as np
```

```python
conn = sqlite3.connect('myDataBase.sqlite')
```

```python
datos = pd.read_csv('Produccin.csv')
```

```python
def tabla_sencilla(datos):
    datos['Potencia'] = pd.to_numeric(datos['Potencia [kW]'],errors = 'ignore')
    datos['Insolacion'] = pd.to_numeric(datos['Insolación [W/m²]'],errors = 'ignore')
    datos = datos.drop(['Potencia [kW]', 'Insolación [W/m²]'],axis= "columns")
    return datos

def datos_limpios(datos):
    datos = datos[datos['Potencia']!='x']
    datos['Potencia']= pd.to_numeric(datos['Potencia'])
    datos['Insolacion']= pd.to_numeric(datos['Insolacion'])
    return datos
```

```python
datos1 = tabla_sencilla(datos)
datos1 = datos_limpios(datos1)
```

```python
datos1.to_sql('Produccion',con=conn,dtype={'time':'TIMESTAMP'})
```
